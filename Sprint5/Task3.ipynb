Описание задачи: 

Создать базовый класс CreditCard в C#, который будет представлять информацию 
о кредитных картах. На основе этого класса разработать 2-3 производных класса, 
демонстрирующих принципы наследования и полиморфизма. В каждом из классов 
должны быть реализованы новые атрибуты и методы, а также переопределены 
некоторые методы базового класса для демонстрации полиморфизма. 
Требования к базовому классу CreditCard: 
• Атрибуты:  Номер  карты  (CardNumber),  Холдера  (HolderName),  Срок 
действия (ExpiryDate). 
• Методы: 
o GetInfo(): метод для получения информации о кредитной карте в виде 
строки. 
o Pay(): метод для оплаты покупки с использованием карты. 
o CheckBalance(): метод для проверки баланса на карте. 
Требования к производным классам: 
1. GoldCard  (GoldCreditCard):  Должен  содержать  дополнительные  атрибуты, 
такие как Бесплатные бонусные мили (BonusMiles). Метод Pay() должен быть 
переопределен для добавления информации о получении бонусных миль 
при оплате покупки. 
2. PremiumCard  (PremiumCreditCard):  Должен  содержать  дополнительные 
атрибуты,  такие  как  Ассистент  поддержки  (SupportAssistant). 
Метод CheckBalance() должен  быть  переопределен  для  предоставления 
возможности  обратиться  за  помощью  к  ассистенту  поддержки  в  случае 
проблем с балансом. 
3. CorporateCard (CorporateCreditCard) (если требуется третий класс): Должен 
содержать  дополнительные  атрибуты,  такие  как  Компания  (Company). 
Метод GetInfo() должен быть переопределен для включения информации о 
компании в описании карты. 


#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) исользуйтие в проекте коллекции, делегаты, события.

```

class CreditCard
{
    public string CardNumber { get; set; }
    public string HolderName { get; set; }
    public string ExpiryDate { get; set; }
    public double Balance { get; set; }

    // Дополнительные атрибуты
    public string BankName { get; set; }
    public string Currency { get; set; }
    public bool IsBlocked { get; set; }
    public double CreditLimit { get; set; }

    // Делегат и событие
    public delegate void CardOperationHandler(string message);
    public event CardOperationHandler CardChanged;

    public CreditCard(string cardNumber, string holderName, string expiryDate, double balance,
                      string bankName, string currency, bool isBlocked, double creditLimit)
    {
        CardNumber = cardNumber;
        HolderName = holderName;
        ExpiryDate = expiryDate;
        Balance = balance;
        BankName = bankName;
        Currency = currency;
        IsBlocked = isBlocked;
        CreditLimit = creditLimit;
    }

    public virtual string GetInfo()
    {
        return $"Номер карты: {CardNumber}, Владелец: {HolderName}, Срок действия: {ExpiryDate}, " +
               $"Банк: {BankName}, Валюта: {Currency}, Лимит: {CreditLimit}";
    }

    public virtual void Pay(double amount)
    {
        if (IsBlocked)
        {
            Console.WriteLine("Карта заблокирована.");
            OnCardChanged($"Попытка оплаты по заблокированной карте {CardNumber}");
            return;
        }

        if (amount <= Balance + CreditLimit)
        {
            Balance -= amount;
            Console.WriteLine($"Оплата на сумму {amount} {Currency} выполнена.");
            OnCardChanged($"С карты {CardNumber} выполнена оплата на {amount} {Currency}");
        }
        else
        {
            Console.WriteLine("Недостаточно средств.");
            OnCardChanged($"Недостаточно средств на карте {CardNumber}");
        }
    }

    public virtual void CheckBalance()
    {
        Console.WriteLine($"Баланс карты: {Balance} {Currency}");
    }

    // Дополнительные методы
    public virtual void TopUp(double amount)
    {
        if (amount > 0)
        {
            Balance += amount;
            Console.WriteLine($"Карта пополнена на {amount} {Currency}");
            OnCardChanged($"Карта {CardNumber} пополнена на {amount} {Currency}");
        }
    }

    public void BlockCard()
    {
        IsBlocked = true;
        Console.WriteLine("Карта заблокирована.");
        OnCardChanged($"Карта {CardNumber} заблокирована");
    }

    public void UnblockCard()
    {
        IsBlocked = false;
        Console.WriteLine("Карта разблокирована.");
        OnCardChanged($"Карта {CardNumber} разблокирована");
    }

    public void ChangeLimit(double newLimit)
    {
        if (newLimit >= 0)
        {
            CreditLimit = newLimit;
            Console.WriteLine($"Новый кредитный лимит: {CreditLimit} {Currency}");
            OnCardChanged($"У карты {CardNumber} изменен лимит на {CreditLimit} {Currency}");
        }
    }

    protected void OnCardChanged(string message)
    {
        CardChanged?.Invoke(message);
    }
}

class GoldCreditCard : CreditCard
{
    public int BonusMiles { get; set; }
    public double CashbackPercent { get; set; }

    public GoldCreditCard(string cardNumber, string holderName, string expiryDate, double balance,
                          string bankName, string currency, bool isBlocked, double creditLimit,
                          int bonusMiles, double cashbackPercent)
        : base(cardNumber, holderName, expiryDate, balance, bankName, currency, isBlocked, creditLimit)
    {
        BonusMiles = bonusMiles;
        CashbackPercent = cashbackPercent;
    }

    public override void Pay(double amount)
    {
        if (IsBlocked)
        {
            Console.WriteLine("Карта заблокирована.");
            return;
        }

        if (amount <= Balance + CreditLimit)
        {
            Balance -= amount;
            BonusMiles += (int)amount / 10;
            double cashback = amount * CashbackPercent / 100.0;
            Balance += cashback;

            Console.WriteLine($"Оплата на сумму {amount} {Currency} выполнена.");
            Console.WriteLine($"Начислены бонусные мили. Всего миль: {BonusMiles}");
            Console.WriteLine($"Начислен кэшбэк: {cashback:F2} {Currency}");
        }
        else
        {
            Console.WriteLine("Недостаточно средств.");
        }
    }

    public void RedeemMiles(int miles)
    {
        if (miles > 0 && miles <= BonusMiles)
        {
            BonusMiles -= miles;
            Console.WriteLine($"Списано {miles} миль. Осталось: {BonusMiles}");
        }
    }
}

class PremiumCreditCard : CreditCard
{
    public string SupportAssistant { get; set; }
    public bool LoungeAccess { get; set; }

    public PremiumCreditCard(string cardNumber, string holderName, string expiryDate, double balance,
                             string bankName, string currency, bool isBlocked, double creditLimit,
                             string supportAssistant, bool loungeAccess)
        : base(cardNumber, holderName, expiryDate, balance, bankName, currency, isBlocked, creditLimit)
    {
        SupportAssistant = supportAssistant;
        LoungeAccess = loungeAccess;
    }

    public override void CheckBalance()
    {
        Console.WriteLine($"Баланс карты: {Balance} {Currency}");
        Console.WriteLine($"При возникновении проблем обратитесь к ассистенту: {SupportAssistant}");
        Console.WriteLine($"Доступ в бизнес-зал: {LoungeAccess}");
    }

    public void ContactAssistant()
    {
        Console.WriteLine($"Связь с ассистентом: {SupportAssistant}");
    }
}

class CorporateCreditCard : CreditCard
{
    public string Company { get; set; }
    public string Department { get; set; }

    public CorporateCreditCard(string cardNumber, string holderName, string expiryDate, double balance,
                               string bankName, string currency, bool isBlocked, double creditLimit,
                               string company, string department)
        : base(cardNumber, holderName, expiryDate, balance, bankName, currency, isBlocked, creditLimit)
    {
        Company = company;
        Department = department;
    }

    public override string GetInfo()
    {
        return $"Номер карты: {CardNumber}, Владелец: {HolderName}, Срок действия: {ExpiryDate}, " +
               $"Компания: {Company}, Отдел: {Department}, Банк: {BankName}";
    }

    public void GenerateReport()
    {
        Console.WriteLine($"Отчет по карте сотрудника {HolderName}, компания {Company}, отдел {Department}");
    }
}

delegate void ShowCardInfo(CreditCard card);

CreditCard card = new CreditCard(
    "1111 2222 3333 4444", "Иван Иванов", "12/27", 50000,
    "СберБанк", "RUB", false, 10000);

GoldCreditCard goldCard = new GoldCreditCard(
    "5555 6666 7777 8888", "Петр Петров", "11/28", 70000,
    "Альфа-Банк", "RUB", false, 15000, 100, 3.0);

PremiumCreditCard premiumCard = new PremiumCreditCard(
    "9999 0000 1111 2222", "Анна Смирнова", "10/29", 120000,
    "Т-Банк", "RUB", false, 30000, "Личный менеджер Алексей", true);

CorporateCreditCard corporateCard = new CorporateCreditCard(
    "3333 4444 5555 6666", "Олег Сидоров", "09/30", 250000,
    "ВТБ", "RUB", false, 50000, "ООО ТехноПром", "Отдел снабжения");

// Подписка на события
card.CardChanged += OnCardChanged;
goldCard.CardChanged += OnCardChanged;
premiumCard.CardChanged += OnCardChanged;
corporateCard.CardChanged += OnCardChanged;

// КОЛЛЕКЦИЯ List<CreditCard>
List<CreditCard> cards = new List<CreditCard>
{
    card,
    goldCard,
    premiumCard,
    corporateCard
};

Console.WriteLine("=== ВСЕ КАРТЫ ===");
foreach (var c in cards)
{
    Console.WriteLine(c.GetInfo());
    Console.WriteLine();
}

// ДЕЛЕГАТ
Console.WriteLine("=== ДЕЛЕГАТ ===");
ShowCardInfo showInfo = ShowInfoMethod;
showInfo(card);
showInfo(goldCard);
Console.WriteLine();

// СОБЫТИЯ + МЕТОДЫ
Console.WriteLine("=== ОПЕРАЦИИ ===");
card.Pay(10000);
card.TopUp(5000);
card.CheckBalance();
Console.WriteLine();

goldCard.Pay(15000);
goldCard.RedeemMiles(50);
goldCard.CheckBalance();
Console.WriteLine();

premiumCard.Pay(20000);
premiumCard.CheckBalance();
premiumCard.ContactAssistant();
Console.WriteLine();

corporateCard.Pay(50000);
corporateCard.CheckBalance();
corporateCard.GenerateReport();
Console.WriteLine();

// КОЛЛЕКЦИЯ Dictionary<string, CreditCard>
Console.WriteLine("=== ПОИСК В Dictionary ===");
Dictionary<string, CreditCard> cardDictionary = new Dictionary<string, CreditCard>();
foreach (var c in cards)
{
    cardDictionary.Add(c.CardNumber, c);
}

if (cardDictionary.TryGetValue("5555 6666 7777 8888", out CreditCard foundCard))
{
    Console.WriteLine("Найдена карта:");
    Console.WriteLine(foundCard.GetInfo());
}

Console.WriteLine();

// ЛЯМБДА-ВЫРАЖЕНИЕ + LINQ
Console.WriteLine("=== КАРТЫ С БАЛАНСОМ БОЛЬШЕ 60000 ===");
var richCards = cards.Where(c => c.Balance > 60000);
foreach (var c in richCards)
{
    Console.WriteLine(c.GetInfo());
}

static void ShowInfoMethod(CreditCard card)
{
Console.WriteLine(card.GetInfo());
}

static void OnCardChanged(string message)
{
Console.WriteLine($"[СОБЫТИЕ] {message}");
}
```
