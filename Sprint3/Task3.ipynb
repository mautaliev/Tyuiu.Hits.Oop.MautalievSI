Описание задачи: 

Создать базовый класс CreditCard в C#, который будет представлять информацию 
о кредитных картах. На основе этого класса разработать 2-3 производных класса, 
демонстрирующих принципы наследования и полиморфизма. В каждом из классов 
должны быть реализованы новые атрибуты и методы, а также переопределены 
некоторые методы базового класса для демонстрации полиморфизма. 
Требования к базовому классу CreditCard: 
• Атрибуты:  Номер  карты  (CardNumber),  Холдера  (HolderName),  Срок 
действия (ExpiryDate). 
• Методы: 
o GetInfo(): метод для получения информации о кредитной карте в виде 
строки. 
o Pay(): метод для оплаты покупки с использованием карты. 
o CheckBalance(): метод для проверки баланса на карте. 
Требования к производным классам: 
1. GoldCard  (GoldCreditCard):  Должен  содержать  дополнительные  атрибуты, 
такие как Бесплатные бонусные мили (BonusMiles). Метод Pay() должен быть 
переопределен для добавления информации о получении бонусных миль 
при оплате покупки. 
2. PremiumCard  (PremiumCreditCard):  Должен  содержать  дополнительные 
атрибуты,  такие  как  Ассистент  поддержки  (SupportAssistant). 
Метод CheckBalance() должен  быть  переопределен  для  предоставления 
возможности  обратиться  за  помощью  к  ассистенту  поддержки  в  случае 
проблем с балансом. 
3. CorporateCard (CorporateCreditCard) (если требуется третий класс): Должен 
содержать  дополнительные  атрибуты,  такие  как  Компания  (Company). 
Метод GetInfo() должен быть переопределен для включения информации о 
компании в описании карты. 

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте полиморфизм с перекрытием и прегегрузкой методов, а также generic классы

```

// Интерфейсы для множественного наследования
interface ITransferable
{
    void Transfer(CreditCard target, double amount);
}
interface IReportable
{
    void GenerateReport();
}

// ===================== БАЗОВЫЙ КЛАСС =====================
class CreditCard
{
    private string _cardNumber;
    private string _holderName;
    private string _expiryDate;
    private double _balance;
    private string _bankName;
    private string _currency;
    private double _creditLimit;
    private bool _isBlocked;

    // Новые атрибуты
    private string _paymentSystem;
    private int _cvv;
    private string _cardStatus;
    private double _serviceFee;

    public string CardNumber
    {
        get { return _cardNumber; }
        set { _cardNumber = value; }
    }

    public string HolderName
    {
        get { return _holderName; }
        set { _holderName = value; }
    }

    public string ExpiryDate
    {
        get { return _expiryDate; }
        set { _expiryDate = value; }
    }

    public double Balance
    {
        get { return _balance; }
        set
        {
            _balance = value;
        }
    }

    public string BankName
    {
        get { return _bankName; }
        set { _bankName = value; }
    }

    public string Currency
    {
        get { return _currency; }
        set { _currency = value; }
    }

    public double CreditLimit
    {
        get { return _creditLimit; }
        set
        {
            if (value >= 0)
                _creditLimit = value;
        }
    }

    public bool IsBlocked
    {
        get { return _isBlocked; }
        set { _isBlocked = value; }
    }

    public string PaymentSystem
    {
        get { return _paymentSystem; }
        set { _paymentSystem = value; }
    }

    public int Cvv
    {
        get { return _cvv; }
        set
        {
            if (value >= 0 && value <= 999)
                _cvv = value;
        }
    }

    public string CardStatus
    {
        get { return _cardStatus; }
        set { _cardStatus = value; }
    }

    public double ServiceFee
    {
        get { return _serviceFee; }
        set
        {
            if (value >= 0)
                _serviceFee = value;
        }
    }

    public CreditCard(
        string cardNumber,
        string holderName,
        string expiryDate,
        double balance,
        string bankName,
        string currency,
        double creditLimit,
        bool isBlocked,
        string paymentSystem,
        int cvv,
        string cardStatus,
        double serviceFee)
    {
        CardNumber = cardNumber;
        HolderName = holderName;
        ExpiryDate = expiryDate;
        Balance = balance;
        BankName = bankName;
        Currency = currency;
        CreditLimit = creditLimit;
        IsBlocked = isBlocked;
        PaymentSystem = paymentSystem;
        Cvv = cvv;
        CardStatus = cardStatus;
        ServiceFee = serviceFee;
    }

    public virtual string GetInfo()
    {
        return $"Номер: {CardNumber}, Владелец: {HolderName}, Срок: {ExpiryDate}, " +
               $"Банк: {BankName}, Валюта: {Currency}, Лимит: {CreditLimit}, " +
               $"Платежная система: {PaymentSystem}, Статус: {CardStatus}, Комиссия: {ServiceFee}";
    }

    // Перекрываемый метод
    public virtual void Pay(double amount)
    {
        if (IsBlocked)
        {
            Console.WriteLine("Карта заблокирована. Оплата невозможна.");
            return;
        }

        if (amount <= Balance + CreditLimit)
        {
            Balance -= amount;
            Console.WriteLine($"{HolderName} оплатил {amount} {Currency}");
        }
        else
        {
            Console.WriteLine("Недостаточно средств");
        }
    }

    // Перегрузка метода Pay
    public virtual void Pay(double amount, string shopName)
    {
        if (IsBlocked)
        {
            Console.WriteLine("Карта заблокирована. Оплата невозможна.");
            return;
        }

        if (amount <= Balance + CreditLimit)
        {
            Balance -= amount;
            Console.WriteLine($"{HolderName} оплатил {amount} {Currency} в магазине {shopName}");
        }
        else
        {
            Console.WriteLine("Недостаточно средств");
        }
    }

    public virtual void CheckBalance()
    {
        Console.WriteLine($"Баланс карты {HolderName}: {Balance} {Currency}");
    }

    public void TopUp(double amount)
    {
        if (amount > 0)
        {
            Balance += amount;
            Console.WriteLine($"Карта {HolderName} пополнена на {amount} {Currency}");
        }
    }

    // Перегрузка TopUp
    public void TopUp(double amount, string source)
    {
        if (amount > 0)
        {
            Balance += amount;
            Console.WriteLine($"Карта {HolderName} пополнена на {amount} {Currency} через {source}");
        }
    }

    public void BlockCard()
    {
        IsBlocked = true;
        CardStatus = "Заблокирована";
        Console.WriteLine($"Карта {HolderName} заблокирована");
    }

    public void UnblockCard()
    {
        IsBlocked = false;
        CardStatus = "Активна";
        Console.WriteLine($"Карта {HolderName} разблокирована");
    }

    public void WithdrawCash(double amount)
    {
        if (IsBlocked)
        {
            Console.WriteLine("Карта заблокирована. Снятие невозможно.");
            return;
        }

        if (amount <= Balance)
        {
            Balance -= amount;
            Console.WriteLine($"{HolderName} снял {amount} {Currency}");
        }
        else
        {
            Console.WriteLine("Недостаточно средств для снятия");
        }
    }

    // Новый метод
    public void ApplyServiceFee()
    {
        Balance -= ServiceFee;
        Console.WriteLine($"С карты {HolderName} списана комиссия {ServiceFee} {Currency}");
    }

    // Новый метод
    public void ChangeStatus(string newStatus)
    {
        CardStatus = newStatus;
        Console.WriteLine($"Статус карты {HolderName} изменен на: {CardStatus}");
    }
}

// ===================== ПРОСТОЕ НАСЛЕДОВАНИЕ =====================
class GoldCreditCard : CreditCard
{
    public int BonusMiles { get; set; }
    public double CashbackPercent { get; set; }
    public string LoungeAccess { get; set; }

    // Новые атрибуты
    public int FreeTransfersPerMonth { get; set; }
    public string PersonalManager { get; set; }

    public GoldCreditCard(
        string cardNumber,
        string holderName,
        string expiryDate,
        double balance,
        string bankName,
        string currency,
        double creditLimit,
        bool isBlocked,
        string paymentSystem,
        int cvv,
        string cardStatus,
        double serviceFee,
        int bonusMiles,
        double cashbackPercent,
        string loungeAccess,
        int freeTransfersPerMonth,
        string personalManager)
        : base(cardNumber, holderName, expiryDate, balance, bankName, currency, creditLimit,
               isBlocked, paymentSystem, cvv, cardStatus, serviceFee)
    {
        BonusMiles = bonusMiles;
        CashbackPercent = cashbackPercent;
        LoungeAccess = loungeAccess;
        FreeTransfersPerMonth = freeTransfersPerMonth;
        PersonalManager = personalManager;
    }

    // Перекрытие
    public override void Pay(double amount)
    {
        if (IsBlocked)
        {
            Console.WriteLine("Gold-карта заблокирована. Оплата невозможна.");
            return;
        }

        if (amount <= Balance + CreditLimit)
        {
            Balance -= amount;
            BonusMiles += (int)(amount / 10);
            double cashback = amount * CashbackPercent / 100.0;

            Console.WriteLine($"Gold-оплата: {amount} {Currency}");
            Console.WriteLine($"Начислено миль: {BonusMiles}");
            Console.WriteLine($"Кэшбэк: {cashback} {Currency}");
        }
        else
        {
            Console.WriteLine("Недостаточно средств");
        }
    }

    // Перекрытие + перегрузка
    public override void Pay(double amount, string shopName)
    {
        if (IsBlocked)
        {
            Console.WriteLine("Gold-карта заблокирована. Оплата невозможна.");
            return;
        }

        if (amount <= Balance + CreditLimit)
        {
            Balance -= amount;
            BonusMiles += (int)(amount / 10);
            Console.WriteLine($"Gold-карта {HolderName} оплатила {amount} {Currency} в магазине {shopName}");
            Console.WriteLine($"Новые бонусные мили: {BonusMiles}");
        }
        else
        {
            Console.WriteLine("Недостаточно средств");
        }
    }

    public void UseLounge()
    {
        Console.WriteLine($"{HolderName} использует доступ в бизнес-зал: {LoungeAccess}");
    }

    public void ContactManager()
    {
        Console.WriteLine($"{HolderName} обращается к персональному менеджеру: {PersonalManager}");
    }

    public override string GetInfo()
    {
        return base.GetInfo() +
               $", Бонусные мили: {BonusMiles}, Кэшбэк: {CashbackPercent}%, " +
               $"Lounge: {LoungeAccess}, Бесплатных переводов: {FreeTransfersPerMonth}, Менеджер: {PersonalManager}";
    }
}

// ===================== СЛОЖНОЕ НАСЛЕДОВАНИЕ =====================
class PremiumCreditCard : GoldCreditCard
{
    public string SupportAssistant { get; set; }
    public string InsurancePackage { get; set; }
    public int PriorityLevel { get; set; }

    // Новые атрибуты
    public bool ConciergeService { get; set; }
    public string VipStatus { get; set; }

    public PremiumCreditCard(
        string cardNumber,
        string holderName,
        string expiryDate,
        double balance,
        string bankName,
        string currency,
        double creditLimit,
        bool isBlocked,
        string paymentSystem,
        int cvv,
        string cardStatus,
        double serviceFee,
        int bonusMiles,
        double cashbackPercent,
        string loungeAccess,
        int freeTransfersPerMonth,
        string personalManager,
        string supportAssistant,
        string insurancePackage,
        int priorityLevel,
        bool conciergeService,
        string vipStatus)
        : base(cardNumber, holderName, expiryDate, balance, bankName, currency,
               creditLimit, isBlocked, paymentSystem, cvv, cardStatus, serviceFee,
               bonusMiles, cashbackPercent, loungeAccess, freeTransfersPerMonth, personalManager)
    {
        SupportAssistant = supportAssistant;
        InsurancePackage = insurancePackage;
        PriorityLevel = priorityLevel;
        ConciergeService = conciergeService;
        VipStatus = vipStatus;
    }

    // Перекрытие
    public override void CheckBalance()
    {
        Console.WriteLine($"Premium баланс: {Balance} {Currency}");
        Console.WriteLine($"Персональный ассистент: {SupportAssistant}");
        Console.WriteLine($"Приоритет обслуживания: {PriorityLevel}");
        Console.WriteLine($"VIP-статус: {VipStatus}");
    }

    public void CallAssistant()
    {
        Console.WriteLine($"{HolderName} обращается к ассистенту: {SupportAssistant}");
    }

    public void ActivateInsurance()
    {
        Console.WriteLine($"Для карты {HolderName} подключен страховой пакет: {InsurancePackage}");
    }

    public void UseConcierge()
    {
        if (ConciergeService)
            Console.WriteLine($"{HolderName} использует услугу консьерж-сервиса");
        else
            Console.WriteLine("Консьерж-сервис недоступен");
    }

    public override string GetInfo()
    {
        return base.GetInfo() +
               $", Ассистент: {SupportAssistant}, Страховка: {InsurancePackage}, " +
               $"Приоритет: {PriorityLevel}, Concierge: {ConciergeService}, VIP: {VipStatus}";
    }
}

// ===================== МНОЖЕСТВЕННОЕ НАСЛЕДОВАНИЕ ЧЕРЕЗ ИНТЕРФЕЙСЫ =====================
class CorporateCreditCard : CreditCard, ITransferable, IReportable
{
    public string Company { get; set; }
    public string Department { get; set; }
    public double SpendingLimitPerMonth { get; set; }

    // Новые атрибуты
    public string Position { get; set; }
    public int ApprovedTransactionsCount { get; set; }

    public CorporateCreditCard(
        string cardNumber,
        string holderName,
        string expiryDate,
        double balance,
        string bankName,
        string currency,
        double creditLimit,
        bool isBlocked,
        string paymentSystem,
        int cvv,
        string cardStatus,
        double serviceFee,
        string company,
        string department,
        double spendingLimitPerMonth,
        string position,
        int approvedTransactionsCount)
        : base(cardNumber, holderName, expiryDate, balance, bankName, currency, creditLimit,
               isBlocked, paymentSystem, cvv, cardStatus, serviceFee)
    {
        Company = company;
        Department = department;
        SpendingLimitPerMonth = spendingLimitPerMonth;
        Position = position;
        ApprovedTransactionsCount = approvedTransactionsCount;
    }

    public override string GetInfo()
    {
        return base.GetInfo() +
               $", Компания: {Company}, Отдел: {Department}, Лимит в месяц: {SpendingLimitPerMonth}, " +
               $"Должность: {Position}, Подтвержденных операций: {ApprovedTransactionsCount}";
    }

    // Перекрытие
    public override void Pay(double amount)
    {
        if (IsBlocked)
        {
            Console.WriteLine("Корпоративная карта заблокирована. Оплата невозможна.");
            return;
        }

        if (amount <= Balance && amount <= SpendingLimitPerMonth)
        {
            Balance -= amount;
            ApprovedTransactionsCount++;
            Console.WriteLine($"Корпоративная карта оплатила {amount} {Currency}");
        }
        else
        {
            Console.WriteLine("Превышен баланс или месячный лимит");
        }
    }

    public void Transfer(CreditCard target, double amount)
    {
        if (IsBlocked)
        {
            Console.WriteLine("Корпоративная карта заблокирована. Перевод невозможен.");
            return;
        }

        if (amount <= Balance)
        {
            Balance -= amount;
            target.Balance += amount;
            Console.WriteLine($"{HolderName} перевел {amount} {Currency} -> {target.HolderName}");
        }
        else
        {
            Console.WriteLine("Недостаточно средств для перевода");
        }
    }

    // Перегрузка Transfer
    public void Transfer(CreditCard target, double amount, string comment)
    {
        if (IsBlocked)
        {
            Console.WriteLine("Корпоративная карта заблокирована. Перевод невозможен.");
            return;
        }

        if (amount <= Balance)
        {
            Balance -= amount;
            target.Balance += amount;
            Console.WriteLine($"{HolderName} перевел {amount} {Currency} -> {target.HolderName}. Комментарий: {comment}");
        }
        else
        {
            Console.WriteLine("Недостаточно средств для перевода");
        }
    }

    public void GenerateReport()
    {
        Console.WriteLine("----- ОТЧЕТ ПО КОРПОРАТИВНОЙ КАРТЕ -----");
        Console.WriteLine($"Компания: {Company}");
        Console.WriteLine($"Сотрудник: {HolderName}");
        Console.WriteLine($"Отдел: {Department}");
        Console.WriteLine($"Должность: {Position}");
        Console.WriteLine($"Баланс: {Balance} {Currency}");
        Console.WriteLine($"Месячный лимит: {SpendingLimitPerMonth} {Currency}");
        Console.WriteLine($"Подтвержденных операций: {ApprovedTransactionsCount}");
    }

    public void PayForOffice(double amount)
    {
        Console.WriteLine($"{HolderName} оплатил офисные расходы на {amount} {Currency}");
        Pay(amount);
    }

    public void ApproveTransaction()
    {
        ApprovedTransactionsCount++;
        Console.WriteLine($"Операция одобрена. Всего подтверждено: {ApprovedTransactionsCount}");
    }
}

// ===================== GENERIC-КЛАСС =====================
class CardCollection<T> where T : CreditCard
{
    private List<T> cards = new List<T>();

    public void AddCard(T card)
    {
        cards.Add(card);
    }

    public void ShowAllCards()
    {
        foreach (T card in cards)
        {
            Console.WriteLine(card.GetInfo());
        }
    }

    public void CheckAllBalances()
    {
        foreach (T card in cards)
        {
            card.CheckBalance();
        }
    }
}


CreditCard card = new CreditCard(
    "1111 2222 3333 4444", "Иван Иванов", "12/27",
    50000, "SberBank", "RUB", 10000, false,
    "Visa", 123, "Активна", 150);

GoldCreditCard gold = new GoldCreditCard(
    "2222 3333 4444 5555", "Петр Петров", "11/28",
    70000, "Tinkoff", "RUB", 20000, false,
    "MasterCard", 456, "Активна", 200,
    100, 5.0, "Доступ открыт", 3, "Сергей");

PremiumCreditCard premium = new PremiumCreditCard(
    "3333 4444 5555 6666", "Анна Смирнова", "10/29",
    120000, "AlfaBank", "RUB", 50000, false,
    "Visa", 789, "Активна", 300,
    300, 7.0, "VIP Lounge", 5, "Дмитрий",
    "Алексей", "Полная страховка", 1, true, "Platinum");

CorporateCreditCard corporate = new CorporateCreditCard(
    "4444 5555 6666 7777", "Олег Сидоров", "09/30",
    200000, "VTB", "RUB", 100000, false,
    "МИР", 321, "Активна", 500,
    "TechCorp", "IT", 300000, "Руководитель отдела", 10);

Console.WriteLine("=== Информация о картах ===");
Console.WriteLine(card.GetInfo());
Console.WriteLine(gold.GetInfo());
Console.WriteLine(premium.GetInfo());
Console.WriteLine(corporate.GetInfo());

Console.WriteLine("\n=== Полиморфизм с перекрытием ===");
CreditCard[] cardsArray = { card, gold, premium, corporate };
foreach (CreditCard item in cardsArray)
{
    item.CheckBalance();
}

Console.WriteLine("\n=== Перегрузка методов ===");
card.Pay(5000);
card.Pay(3000, "DNS");
gold.Pay(10000);
gold.Pay(7000, "M.Video");
card.TopUp(2000);
card.TopUp(1500, "банкомат");
corporate.Transfer(card, 10000);
corporate.Transfer(gold, 5000, "Премия сотруднику");

Console.WriteLine("\n=== Дополнительные методы ===");
card.ApplyServiceFee();
card.ChangeStatus("На проверке");
gold.UseLounge();
gold.ContactManager();
premium.CallAssistant();
premium.ActivateInsurance();
premium.UseConcierge();
corporate.PayForOffice(25000);
corporate.ApproveTransaction();

Console.WriteLine("\n=== Generic-класс ===");
CardCollection<CreditCard> allCards = new CardCollection<CreditCard>();
allCards.AddCard(card);
allCards.AddCard(gold);
allCards.AddCard(premium);
allCards.AddCard(corporate);

allCards.ShowAllCards();

Console.WriteLine("\n=== Проверка балансов через Generic-класс ===");
allCards.CheckAllBalances();

Console.WriteLine("\n=== Отчет ===");
corporate.GenerateReport();

Console.WriteLine("\n=== Блокировка карты ===");
card.BlockCard();
card.Pay(1000);
card.UnblockCard();
card.Pay(1000);
```
