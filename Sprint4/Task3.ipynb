Описание задачи: 

Создать базовый класс CreditCard в C#, который будет представлять информацию 
о кредитных картах. На основе этого класса разработать 2-3 производных класса, 
демонстрирующих принципы наследования и полиморфизма. В каждом из классов 
должны быть реализованы новые атрибуты и методы, а также переопределены 
некоторые методы базового класса для демонстрации полиморфизма. 
Требования к базовому классу CreditCard: 
• Атрибуты:  Номер  карты  (CardNumber),  Холдера  (HolderName),  Срок 
действия (ExpiryDate). 
• Методы: 
o GetInfo(): метод для получения информации о кредитной карте в виде 
строки. 
o Pay(): метод для оплаты покупки с использованием карты. 
o CheckBalance(): метод для проверки баланса на карте. 
Требования к производным классам: 
1. GoldCard  (GoldCreditCard):  Должен  содержать  дополнительные  атрибуты, 
такие как Бесплатные бонусные мили (BonusMiles). Метод Pay() должен быть 
переопределен для добавления информации о получении бонусных миль 
при оплате покупки. 
2. PremiumCard  (PremiumCreditCard):  Должен  содержать  дополнительные 
атрибуты,  такие  как  Ассистент  поддержки  (SupportAssistant). 
Метод CheckBalance() должен  быть  переопределен  для  предоставления 
возможности  обратиться  за  помощью  к  ассистенту  поддержки  в  случае 
проблем с балансом. 
3. CorporateCard (CorporateCreditCard) (если требуется третий класс): Должен 
содержать  дополнительные  атрибуты,  такие  как  Компания  (Company). 
Метод GetInfo() должен быть переопределен для включения информации о 
компании в описании карты.

```
// В спринтах 2 и 3 чет затупил и получилось так, что в спринте 2 усложнил код из спринта 1
// в спринте 3 - из спринта 2
// а как будто бы везде за основу надо было брать просто спринт 0 (сейчас сделал так)
// а то если и здесь за основу брать спринт 3, то получается слишком много


// ИНТЕРФЕЙСЫ

public interface ICreditCard
{
    string GetInfo();
    void Pay(double amount);
    void CheckBalance();
    void TopUp(double amount);
}

public interface ICardService
{
    void ProcessPayment(CreditCard card, double amount);
    void ProcessTopUp(CreditCard card, double amount);
    void ShowBalance(CreditCard card);
}

// СЕРВИС

public class CardService : ICardService
{
    public void ProcessPayment(CreditCard card, double amount)
    {
        Console.WriteLine($"[Сервис] Обработка оплаты...");
        card.Pay(amount);
    }

    public void ProcessTopUp(CreditCard card, double amount)
    {
        Console.WriteLine($"[Сервис] Пополнение карты...");
        card.TopUp(amount);
    }

    public void ShowBalance(CreditCard card)
    {
        Console.WriteLine($"[Сервис] Проверка баланса...");
        card.CheckBalance();
    }
}

// БАЗОВЫЙ КЛАСС

public class CreditCard : ICreditCard
{
    public string CardNumber { get; set; }
    public string HolderName { get; set; }
    public string ExpiryDate { get; set; }
    public double Balance { get; set; }

    // Доп. атрибуты
    public string BankName { get; set; }
    public string Currency { get; set; }
    public bool IsBlocked { get; set; }
    public double DailyLimit { get; set; }

    public CreditCard(string cardNumber, string holderName, string expiryDate,
                        double balance, string bankName, string currency,
                        bool isBlocked, double dailyLimit)
    {
        CardNumber = cardNumber;
        HolderName = holderName;
        ExpiryDate = expiryDate;
        Balance = balance;
        BankName = bankName;
        Currency = currency;
        IsBlocked = isBlocked;
        DailyLimit = dailyLimit;
    }

    public virtual string GetInfo()
    {
        return $"Карта: {CardNumber}, Владелец: {HolderName}, Банк: {BankName}";
    }

    public virtual void Pay(double amount)
    {
        if (IsBlocked)
        {
            Console.WriteLine("Карта заблокирована.");
            return;
        }

        if (amount > DailyLimit)
        {
            Console.WriteLine("Превышен дневной лимит.");
            return;
        }

        if (amount <= Balance)
        {
            Balance -= amount;
            Console.WriteLine($"Оплачено {amount} {Currency}");
        }
        else
        {
            Console.WriteLine("Недостаточно средств.");
        }
    }

    public virtual void CheckBalance()
    {
        Console.WriteLine($"Баланс: {Balance} {Currency}");
    }

    public virtual void TopUp(double amount)
    {
        Balance += amount;
        Console.WriteLine($"Пополнение на {amount} {Currency}");
    }

    public void BlockCard() => IsBlocked = true;
    public void UnblockCard() => IsBlocked = false;

    // ЯВНАЯ РЕАЛИЗАЦИЯ
    string ICreditCard.GetInfo() => GetInfo();
    void ICreditCard.Pay(double amount) => Pay(amount);
    void ICreditCard.CheckBalance() => CheckBalance();
    void ICreditCard.TopUp(double amount) => TopUp(amount);
}

public class GoldCreditCard : CreditCard
{
    public int BonusMiles { get; set; }

    public GoldCreditCard(string cardNumber, string holderName, string expiryDate,
                            double balance, string bankName, string currency,
                            bool isBlocked, double dailyLimit, int bonusMiles)
        : base(cardNumber, holderName, expiryDate, balance, bankName, currency, isBlocked, dailyLimit)
    {
        BonusMiles = bonusMiles;
    }

    public override void Pay(double amount)
    {
        if (amount <= Balance)
        {
            Balance -= amount;
            int miles = (int)amount / 10;
            BonusMiles += miles;

            Console.WriteLine($"Оплата {amount}. Мили +{miles}");
        }
        else
        {
            Console.WriteLine("Недостаточно средств.");
        }
    }

    public void RedeemMiles(int miles)
    {
        BonusMiles -= miles;
        Console.WriteLine($"Списано миль: {miles}");
    }
}

// PREMIUM

public class PremiumCreditCard : CreditCard
{
    public string SupportAssistant { get; set; }

    public PremiumCreditCard(string cardNumber, string holderName, string expiryDate,
                            double balance, string bankName, string currency,
                            bool isBlocked, double dailyLimit, string assistant)
        : base(cardNumber, holderName, expiryDate, balance, bankName, currency, isBlocked, dailyLimit)
    {
        SupportAssistant = assistant;
    }

    public override void CheckBalance()
    {
        Console.WriteLine($"Баланс: {Balance}");
        Console.WriteLine($"Ассистент: {SupportAssistant}");
    }

    public void ContactAssistant()
    {
        Console.WriteLine($"Связь с: {SupportAssistant}");
    }
}

// CORPORATE

public class CorporateCreditCard : CreditCard
{
    public string Company { get; set; }

    public CorporateCreditCard(string cardNumber, string holderName, string expiryDate,
                                double balance, string bankName, string currency,
                                bool isBlocked, double dailyLimit, string company)
        : base(cardNumber, holderName, expiryDate, balance, bankName, currency, isBlocked, dailyLimit)
    {
        Company = company;
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Компания: {Company}";
    }
}

ICardService service = new CardService();

CreditCard card = new CreditCard("1111", "Иван", "12/27", 50000, "Сбер", "RUB", false, 30000);
GoldCreditCard gold = new GoldCreditCard("2222", "Петр", "11/28", 70000, "Альфа", "RUB", false, 50000, 100);

Console.WriteLine(card.GetInfo());
service.ProcessPayment(card, 10000);
service.ShowBalance(card);

Console.WriteLine();

Console.WriteLine(gold.GetInfo());
service.ProcessPayment(gold, 15000);
service.ShowBalance(gold);
```
